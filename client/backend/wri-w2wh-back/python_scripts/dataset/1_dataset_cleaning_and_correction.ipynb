```python
import pandas as pd
```

### KRS

```python
krs_df = pd.read_csv('reservoir/KRS.csv')
krs_df.head()
```

```python
krs_df['YEAR'].value_counts()
```

```python
krs_df = krs_df[krs_df['YEAR'] != 2010]
```

```python
krs_df['MONTH_DATE'] = krs_df['FLOW_DATE'].str[5:]
krs_df.head()
```

```python
krs_df['MONTH_DATE'].value_counts()
```

```python
krs_df['YEAR'].value_counts()
```

```python
krs_df[krs_df['MONTH_DATE'] == '12-11']
```

```python
krs_df[['WEEK_NO', 'MONTH_DATE']].value_counts().reset_index(name='count')
```

```python
krs_df[krs_df['MONTH_DATE'] == '12-11']
```

```python
krs_df[krs_df['FLOW_DATE'].str[:7] == '2019-12']
```

```python
krs_df[krs_df['SL_NO'] == 23482]
```

```python
krs_df[krs_df['SL_NO'].isin([23481, 23482, 23483, 23484, 23485])]
```

```python
krs_df[krs_df['FLOW_DATE'] >= '2019-11-09'][:6]
```

```python
krs_df.loc[1728, ('FLOW_DATE')] = '2019-11-12'
```

```python
krs_df[krs_df['FLOW_DATE'] >= '2019-11-09'][:6]
```

```python
krs_df = krs_df[['FLOW_DATE', 'RESERVOIR', 'PRESENT_STORAGE_TMC', 'RES_LEVEL_FT', 'INFLOW_CUSECS', 'OUTFLOW_CUECS']]
krs_df
```

```python
krs_df.sort_values(by=['FLOW_DATE'], inplace=True)
krs_df
```

```python
krs_df['FLOW_DATE'].str[:4].value_counts()
```

```python
krs_df.to_csv('reservoir_cleaned/krs.csv', index=False)
```

```python
del krs_df
```
