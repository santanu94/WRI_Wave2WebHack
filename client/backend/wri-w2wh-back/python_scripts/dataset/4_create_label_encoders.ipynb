```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pickle
```

```python
reservoir_df = pd.read_csv('reservoir_dataset.csv')
reservoir_df
```

```python
rainfall_departure_df = pd.read_csv('Rainfall 2010-2020/rainfall_departure_dataset.csv')
rainfall_departure_df
```

```python
month_date_day_df = reservoir_df[['MONTH', 'DATE', 'DAY']].drop_duplicates()
month_date_day_df
```

```python
month_date_day_season_df = month_date_day_df.merge(rainfall_departure_df[['SEASON', 'MONTH']], on='MONTH').drop_duplicates()
month_date_day_season_df
```

```python
month_le = LabelEncoder().fit(month_date_day_season_df['MONTH'].unique())
day_le = LabelEncoder().fit(month_date_day_season_df['DAY'].unique())
season_le = LabelEncoder().fit(month_date_day_season_df['SEASON'].unique())
```

```python
with open('month_le.pkl', 'wb') as f:
    pickle.dump(month_le, f)

with open('day_le.pkl', 'wb') as f:
    pickle.dump(day_le, f)

with open('season_le.pkl', 'wb') as f:
    pickle.dump(season_le, f)

month_date_day_season_df.to_csv('month_date_day_season.csv', index=False)
```
