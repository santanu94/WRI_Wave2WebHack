```python
import pandas as pd
```

```python
aug_inflow_df = pd.read_csv('reservoir_dataset_augmented.csv')
aug_inflow_df
```

```python
aug_inflow_df = aug_inflow_df[aug_inflow_df['RESERVOIR'] == 'K.R.S']
aug_inflow_df
```

```python
aug_inflow_df = aug_inflow_df[aug_inflow_df['YEAR'] != 2021]
aug_inflow_df['YEAR'].value_counts()
```

```python
aug_inflow_df['MONTH_DATE'] = aug_inflow_df['FLOW_DATE'].str[5:]
aug_inflow_df.head()
```

```python
normal_inflow_outflow_df = aug_inflow_df.groupby('MONTH_DATE')[['INFLOW_CUSECS', 'OUTFLOW_CUECS']].mean().reset_index()
normal_inflow_outflow_df
```

```python
normal_inflow_outflow_df.to_csv('datewise_normal_inflow_outflow.csv', index=False)
```
