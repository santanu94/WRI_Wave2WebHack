## Get 2008-2011 data

```python
df1['YEAR'] = df1['DATE'].str[:4]
df1['MONTH'] = df1['DATE'].str[5:7]
df1.head()
```

```python
df1 = df1[(df1['DATE'].str[:4] <= '2011') & (df1['VARIABLE_NAME'] == 'Rainfall (mm.)') & (df1['INDIAN_STATES_NAME'] == 'Karnataka')].sort_values(by='DATE')
df1
```

```python
df1 = df1.pivot(index=['DISTRICTS_NAME', 'YEAR'], columns='MONTH', values='VALUE').reset_index()
df1
```

```python
df1.rename(columns={'DISTRICTS_NAME': 'DISTRICTS',
                    '01': 'JAN',
                    '02': 'FEB',
                    '03': 'MAR',
                    '04': 'APR',
                    '05': 'MAY',
                    '06': 'JUN',
                    '07': 'JUL',
                    '08': 'AUG',
                    '09': 'SEP',
                    '10': 'OCT',
                    '11': 'NOV',
                    '12': 'DEC'}, inplace=True)
df1
```

```python
df_2012.info()
```

```python
df1['DISTRICTS'] = df1['DISTRICTS'].str.upper()
df1
```

## Data Corrections

```python
pair_list = [('CHIKBALLAPURA', 'CHICKBALLAPUR'), ('MYSORE', 'MYSURU'),
             ('UTTAR KANNADA', 'UTTARA KANNADA'), ('BANGALORE RURAL', 'BENGALURU RURAL'),
             ('BANGALORE URBAN', 'BENGALURU URBAN'), ('DAKSHIN KANNADA', 'DAKSHINA KANNADA'),
             ('CHIKMAGALUR', 'CHIKKAMAGALURU'), ('SHIMOGA', 'SHIVAMOGGA'), ('BELGAUM', 'BELAGAVI'),
             ('BIJAPUR', 'VIJAYAPURA'), ('GULBARGA', 'KALABURGI')]

for to_replace, value in pair_list:
    print(to_replace, '->', value)
    df1['DISTRICTS'].replace(to_replace=to_replace, value=value, inplace=True)
```

```python
df1.to_csv('Rainfall-2008-2011.csv', index=False)
```
