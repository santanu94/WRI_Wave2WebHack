```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
```

```python
df = pd.read_csv('opwnweathermap_historic.csv')
df
```

```python
df['dt_iso'].str.split().str[-1].unique()
```

```python
df['date'] = df['dt_iso'].str.split().str[0]
df['time'] = df['dt_iso'].str.split().str[1]
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['dt', 'dt_iso', 'timezone', 'lat', 'lon', 'sea_level', 'grnd_level', 'rain_3h', 'snow_1h', 'snow_3h', 'weather_id', 'weather_icon'], inplace=True)
df.head()
```

```python
df['rain_1h'].fillna(0, inplace=True)
df.head()
```

```python
df['city_name'].unique()
```

```python
kodagu_df = df[df['city_name'] == 'Kodagu']
mysuru_df = df[df['city_name'] == 'Mysuru']
hassan_df = df[df['city_name'] == 'Hassan']

kodagu_df.shape, mysuru_df.shape, hassan_df.shape
```

```python
df['weather_description'].unique()
```

```python
daily_data = {}
for group_key, group_df in df.groupby(['date', 'city_name']):
#     print(group_key)
#     print(group_df)
    date = group_key[0]
    city = group_key[1]
    
    row_key = f'{date} {city}'
    
    daily_data[row_key] = {}
    daily_data[row_key]['date'] = date
    daily_data[row_key]['city_name'] = city
    daily_data[row_key]['temp'] = round(group_df['temp'].mean(), 2)
    daily_data[row_key]['feels_like'] = round(group_df['feels_like'].mean(), 2)
    daily_data[row_key]['temp_min'] = round(group_df['feels_like'].min(), 2)
    daily_data[row_key]['temp_max'] = round(group_df['feels_like'].max(), 2)
    daily_data[row_key]['pressure'] = round(group_df['pressure'].mean(), 2)
    daily_data[row_key]['humidity'] = round(group_df['humidity'].mean(), 2)
    daily_data[row_key]['wind_speed'] = round(group_df['wind_speed'].mean(), 2)
    daily_data[row_key]['wind_deg'] = round(group_df['wind_deg'].mean(), 2)
    daily_data[row_key]['rain'] = round(group_df['rain_1h'].sum(), 2)
    daily_data[row_key]['clouds_all'] = round(group_df['clouds_all'].mean(), 2)
    
    weather_col = group_df['weather_description'].values
    weathers, counts = np.unique(weather_col, return_counts=True)
    for i in range(weathers.shape[0]):
        daily_data[row_key][weathers[i]] = int(counts[i])
#     print(daily_data)
```

```python
daily_df = pd.DataFrame(daily_data)
daily_df
```

```python
daily_df = daily_df.T.reset_index(drop=True)
daily_df
```

```python
daily_df.fillna(0, inplace=True)
```

```python
daily_df[daily_df.columns.difference(['date', 'city_name', 'temp', 'feels_like', 'temp_min', 'temp_max', 'pressure', 'humidity', 'wind_speed', 'wind_deg', 'rain', 'clouds_all'])].sum()
```

```python
daily_df.to_csv('mysuru_kodagu_hassan_weather.csv', index=False)
```
